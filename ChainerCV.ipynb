```python

import numpy as np
import matplotlib.pyplot as plt
from chainercv.datasets import voc_bbox_label_names

from chainercv.links import SSD300
from chainercv.utils import read_image
from chainercv.visualizations import vis_bbox
```

```python
#tasks ChainerCv performs
#Visualization,BboxDataset,Detection Link,DetectionEvaluator,Training script for various detection models
from chainercv.links import SSD300

img = read_image('sample.jpg')

model = SSD300(pretrained_model ='voc0712')
bboxes, labels, scores = model.predict([img])
vis_bbox( img, bboxes[0], labels[0], scores[0], label_names = voc_bbox_label_names)
plt.show()
```
